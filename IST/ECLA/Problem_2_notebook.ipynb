# Laser pulses - Problem 2

```python
import numpy as np
import matplotlib.pyplot as plt

# Define space coordinates
x = np.linspace(0, 100, 500)

# Define time coordinates
t = np.linspace(0, 49.9, 500)
```

```python
# Compute E field of monochromatic wave of frequency omega
omega = 0.5
k = omega
E_0 = 1 # amplitude of the wave
timestep_plot = 1 # timestep to plot

E = E_0 * np.exp(1j*(k*x-omega*t[timestep_plot]))

# Plot E(x) at a given t
plt.plot(x,np.real(E))
plt.xlabel('x')
plt.ylabel('E')
plt.title('t='+'{:.2f}'.format((t[timestep_plot])))
plt.grid()
```
